```python
# using python lists

import time


size = 10_00_000

l1 =list (range(size))
l2 =list (range(size))

start = time.time()

add = [x+y for x,y in zip(l1 , l2)]

end = time.time()

print(f"{end - start} :- This is time taken to compile the code using python list")
```

```python
# using numpy arrays

import numpy as np
import time


size = 10_00_000_0

l1 =  np.array(list(range(size)))
l2 =  np.array(list(range(size)))

start = time.time()

add = [x+y for x,y in zip(l1 , l2)]

end = time.time()

print(f"{end - start} :- This is time taken to compile the code using numpy arrays")
```
