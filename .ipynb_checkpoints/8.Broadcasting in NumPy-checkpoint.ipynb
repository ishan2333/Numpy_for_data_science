<h3>Broadcasting in NumPy</h3>

<p>Loops uses python's Interpreter.Thus they are slow. To make things faster we will use Concept of <b>Vectorization</b> and <b>Broadcasting</b></p>

<b>Vectorization</b> (fixing Loop Problem)

```python
arr = np.array([1, 2, 3, 4, 5])
result = arr ** 2 # Vectorized operation
print(result) # Output: [1 4 9 16 25]
```

<b>Broadcasting</b> <p>is a powerful feature of NumPy that allows you to perform
operations on arrays of different shapes without creating copies.</p>

```python
arr = np.array([1, 2, 3, 4, 5])
result = arr + 10 # Broadcasting: 10 is added to all elements
print(result) # Output: [11 12 13 14 15]
```

<b>Broadcasting with Arrays of Different Shapes</b>

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([10, 20, 30])
result = arr1 + arr2  # Element-wise addition
print(result) # Output: [11 22 33]
```

<h3>Hands-on: Applying Broadcasting to Real-World Scenarios</h3>

```python
#Simulating a dataset (5 samples, 3 features)
data = np.array([[10, 20, 30],
[15, 25, 35],
[20, 30, 40],
[25, 35, 45],
[30, 40, 50]])
# Calculating mean and standard deviation for each feature (column)
mean = data.mean(axis=0)
std = data.std(axis=0)
# Normalizing the data using broadcasting
normalized_data = (data - mean) / std
print(normalized_data)
```
