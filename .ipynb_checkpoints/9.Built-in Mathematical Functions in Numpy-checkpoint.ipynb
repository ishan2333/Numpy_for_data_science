<h3>Built in Mathematical Functions in NumPy</h3>

```python
import numpy as np
```

```python
arr = np.array([3,4,6,8,3,6,8,445,53,73,83,2,4,72,2,42,523,352,23,46,8645,35])
arr2 = np.array([83,2,4,72,2,42,523,352,23,46,8])
```

```python
np.mean(arr)
```

```python
np.std(arr)
```

```python
np.var(arr)
```

```python
np.min(arr)
```

```python
np.max(arr)
```

```python
np.sum(arr)
```

```python
np.prod(arr)
```

```python
np.median(arr)
```

```python
np.percentile(arr, 50) 
```

```python
np.argmin(arr)
```

```python
np.argmax(arr)
```

```python
np.corrcoef(arr, arr2)
```

```python
np.unique(arr)
```

```python
np.diff(arr)
```

```python
np.cumsum(arr)
```

```python
np.linspace(0, 10, 5)
```

```python
np.log(arr)
```

```python
np.exp(arr)
```
