NumPy arrays are <b>homogeneous</b>, meaning that they can <b>only store elements of the
same type.</b>

Common Data Types in NumPy:<br>

int32 , int64 : Integer types with different bit sizes.<br>
float32 , float64 : Floating-point types with different precision.<br>
bool : Boolean data type.<br>
complex64 , complex128 : Complex number types.<br>
object : For storing objects (e.g., Python objects, strings).

You can check the dtype of a NumPy array using the <b>.dtype</b> attribute.


```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print(arr.dtype) # Output: int64 (or int32 depending on the system)
```

<h3>Changing Data Types</h3
<p>You can cast (convert) the data type of an array using the .astype() method. This
is useful when you need to change the type for a specific operation or when you
want to reduce memory usage</p>

```python
arr = np.array([1.5, 2.7, 3.9])
print(arr.dtype) # Output: float64

arr_int = arr.astype(np.int32) # Converting float to int
print(arr_int)
# Output: [1 2 3]


print(arr_int.dtype) # Output: int32
```

<h3>Downcasting</h3>
<p>Downcasting is a way to save a memory consumption. It typecastes data to bigger data type to smaller data type</p>

```python
arr_large = np.array([1000000, 2000000, 3000000], dtype=np.int64)
arr_small = arr_large.astype(np.int32) # Downcasting to a smaller dtype


print(arr_small) # Output: [1000000 2000000 3000000]
print(arr_small.dtype) # Output: int32
```

<p>Following examples states memory consumption by the same data. But Choosing the appropriate data type can save the memory consumption</p>

```python
arr_int64 = np.array([1, 2, 3], dtype=np.int64)
arr_int32 = np.array([1, 2, 3], dtype=np.int32)
print(arr_int64.nbytes) # Output: 24 bytes (3 elements * 8 bytes each)
print(arr_int32.nbytes) # Output: 12 bytes (3 elements * 4 bytes each)
```

<b>String Data Type in NumPy</b> <br>(String are not prefered to store in numpy array)

```python
arr = np.array(['apple', 'banana', 'cherry'], dtype='U10') # Unicode string array ('10' is the size of the string)
arr2 = np.array(['apple', 'banana', 'cherry'], dtype='str') # Unicode string array (automatically declares size)

print(arr.dtype)
print(arr2.dtype)
```

<b>Complex Numbers</b>

```python
arr = np.array([1 + 2j, 3 + 4j, 5 + 6j], dtype='complex128')
print(arr)
```

<b>Object Data Type</b> <br>(objects are not prefered to store in numpy array)

```python
arr = np.array([{'a': 1}, [1, 2, 3], 'hello'], dtype=object)
print(arr)
```

<h3>Choosing the Right Data Type</h3>
Choosing the correct data type is essential for: <br>- Optimizing memory: Using the smallest data type that fits your data. <br>- Improving performance: Smaller types generally lead to faster operations. <br>- Ensuring precision: Avoid truncating or losing important decimal places or values.
